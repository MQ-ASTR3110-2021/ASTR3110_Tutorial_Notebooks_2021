```python
import numpy as np
import matplotlib.pyplot as plt

def sample_mean_dist(N=10,draws=100):
    dist_mean=np.zeros(draws)
    print(draws)
    print(dist_mean.shape)
    for i in np.arange(draws):
        dist_mean[i]=np.nansum(np.random.uniform(low=0.0, high=10.0, size=N))
    return dist_mean
```

```python
plt.clf()
plt.hist(sample_mean_dist(N=100,draws=20000))
plt.show()
```
